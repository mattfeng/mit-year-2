```python
derived["PopP"] = derived.iloc[:, 8] / (120. + 120. + 180.)
```

```python
d56 = (derived.iloc[:, 5] + derived.iloc[:, 6])
d57 = (derived.iloc[:, 5] + derived.iloc[:, 7])
```

```python
d_5_6 = abs(derived.iloc[:, 5] - derived.iloc[:, 6]) / d56 / 2.
```

```python
d56p = d56 / 240.
```

```python
H56 = d_5_6**2 / (d56p * (1 - d56p))
```

```python
d_5_7 = abs(derived.iloc[:, 5] - derived.iloc[:, 7]) / d57 / 2.
```

```python
d57p = d57 / 300
H57 = d_5_7**2 / (d57p * (1 - d57p))
```

```python
H = 0.5 * (H56 + H57)
```

```python
derived[(H > 0.6) & hi_xpehh & high_daf]
```
